```python
import pandas as pd
```

```python
file = pd.read_csv('F:\\datasets\\mushrooms.csv')
df = file.copy()
df.sample(10)
```

```python
df.isnull().sum()
```

```python
for col in df.columns :
    print(col)
    print(len(set(df[col])))
```

```python
label = df['class']
df = df.drop(['class'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
lst = []
for i in df.columns :
    x = pd.DataFrame(encoder.fit_transform(df[i]),columns=[i])
    lst.append(x)
encoded_df = pd.concat(lst,axis=1)
encoded_df.info()
```

```python
encoded_df.sample(10)
```

```python
encoded_label = pd.DataFrame(encoder.fit_transform(label),columns=['class'])
```

```python
from sklearn.model_selection import train_test_split
xtr , xts , ytr , yts = train_test_split(encoded_df,encoded_label)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_xtr = pd.DataFrame(scaler.fit_transform(xtr),columns=df.columns)
scaled_xts = pd.DataFrame(scaler.fit_transform(xts),columns=df.columns)
```

```python
from sklearn.svm import LinearSVC
alg = LinearSVC()
alg.fit(scaled_xtr,ytr)
alg.score(scaled_xtr,ytr)*100
```

```python
alg.score(scaled_xts,yts)*100
```

```python
scaled_df = pd.DataFrame(scaler.fit_transform(encoded_df),columns=encoded_df.columns)
alg.fit(scaled_df,encoded_label)
alg.score(scaled_df,encoded_label)*100
```
